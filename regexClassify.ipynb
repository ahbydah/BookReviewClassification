```python
import re
import pandas as pd
import string
import nltk
from pprint import pprint
import sklearn
```

```python
def getData():
    
    data = pd.read_csv("BookReviews.csv", encoding = 'latin1')
    
    data = data.loc[:, ~data.columns.str.contains('^Unnamed')]  #had several (about 4) columns titles "Unnamed"
    transltr = str.maketrans('', '', string.punctuation)
    for i in range(data['text'].size):
        data['text'].iloc[i] = data['text'].iloc[i].lower()
        data['text'].iloc[i] = data['text'].iloc[i].translate(transltr)
        if data['label'].iloc[i] == "i":
            data['label'].iloc[i] = 1
        else: 
            data['label'].iloc[i] = 0
    return data
```

```python
def searchText(textSeries):
    reg1 = r'dnf'
    reg2 = r'g[ai]ve up'
    reg3 = r'(finish|listen|read|dnf)[\w\s]*(chapter|p(age|g)|%|half ?(way)?)'
    reg4 = r'(have|could|did|never|can(not|t))[\w\s]*(finish|g[eo]t (into|through)|go on)'


    reg = reg1 + r'|' + reg2 + r'|'+ reg3 + r'|'+ reg4

    matchDict = {}
    contents = {}
    
    for i in range(textSeries.size):
        sample = textSeries.iloc[i]
        
        matchBool = 0
        contents[i] = re.findall(reg, sample)
        if contents[i]:
            matchBool = 1
        
        matchDict[i] = matchBool
        
    return [pd.Series(matchDict), pd.Series(contents)]
```
