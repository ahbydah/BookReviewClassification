```python
import re
import pandas as pd
import string
import nltk
from pprint import pprint
```

```python
def getData():
    
    data = pd.read_csv("BookReviews.csv", encoding = 'latin1')
    
    data = data.loc[:, ~data.columns.str.contains('^Unnamed')]  #had several (about 4) columns titles "Unnamed"
  
    for i in range(data['text'].size):
        data['text'].iloc[i] = data['text'].iloc[i].lower()
    
    return data
```

```python
literals = [r'(have|could|did|never)(n\'t)? finish', 
            r'dnf', r'(listen|read)[\w\s]*chapter)', r'g[ia]ve up'

quantities = [r'\D\d{1,2}%', r'(page|pg) \d{1,4}', r'\d{1,4} pages?', r'half(way)?']
```

```python
def searchText(textSeries):

   
    regex = r'((listen|read)(?:[\w\s]*)chapter)|(((did|could)(n\'?t| not) finish)|dnf)|gave u'    
    
    
    matchDict = {}
    contents = {}
    
    for i in range(textSeries.size):
        sample = textSeries.iloc[i]
        
        matchBool = 'c'
        contents[i] = re.findall(regex, sample)
        if contents[i]:
            matchBool = 'i'
        
        matchDict[i] = matchBool
        
    return [pd.Series(matchDict), pd.Series(contents)
```

```python
reviewText = getData()

    
reviewText.loc[:, 'classified'], reviewText.loc[:, 'contents'] = searchText(reviewText.loc[:, 'text'])
```
